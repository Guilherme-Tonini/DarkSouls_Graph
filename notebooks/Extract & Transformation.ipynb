```python
def organize(df):
    new_rows = []
    i = 0
    while i < len(df):
        if df.loc[i, 'text'].startswith('#0'):
            action_line = df.loc[i, 'text'][2:].strip()
            id_primeira_fala = None
            npc = df.loc[i, 'npc']

            j = i + 1
            grupo_texto = []
            while j < len(df) and not df.loc[j, 'text'].startswith('#0'):
                if id_primeira_fala is None:
                    id_primeira_fala = df.loc[j, 'id']
                grupo_texto.append(df.loc[j, 'text'])
                j += 1

            if grupo_texto:
                texto_unificado = ' '.join(grupo_texto)
                new_rows.append({
                    'text': texto_unificado,
                    'id': id_primeira_fala,
                    'npc': npc,
                    'Action': action_line
                })

            i = j
        else:
            new_rows.append({
                'text': df.loc[i, 'text'],
                'id': df.loc[i, 'id'],
                'npc': df.loc[i, 'npc'],
                'Action': None
            })
            i += 1

    return pd.DataFrame(new_rows)

```

```python
def organize_other(df):
    info_atual = None
    nova_linha = []

    for i in range(len(df)):
        texto = df.loc[i, 'text']
        npc = df.loc[i, 'npc']
        linha_id = df.loc[i, 'id']

        if texto.startswith('#0'):
            info_atual = texto[2:].strip()
            continue  
        else:
            nova_linha.append({
                'text': texto,
                'id': linha_id,
                'npc': npc,
                'Information': info_atual
            })

    return pd.DataFrame(nova_linha)
```

```python
def extrair_entidades_maiores(texto):
    frases = re.split(r'[.!?;,]', texto)
    entidades = []

    for frase in frases:
        palavras = frase.strip().split()
        n = len(palavras)
        i = 0

        while i < n:
            if i + 4 < n:
                w1, w2, w3, w4, w5 = palavras[i], palavras[i+1].lower(), palavras[i+2], palavras[i+3], palavras[i+4]
                if (w1[0].isupper() and w3[0].isupper() and w4[0].isupper() and w5[0].isupper()
                    and w2 in {'of', 'the'}
                    and all(w.lower() not in stopwords for w in [w1, w3, w4, w5])):
                    entidade = f"{w1} {w2} {w3} {w4} {w5}"
                    entidades.append(entidade)
                    i += 5
                    continue
            
            if i + 3 < n:
                w1, w2, w3, w4 = palavras[i], palavras[i+1].lower(), palavras[i+2], palavras[i+3]
                if (w1[0].isupper() and w3[0].isupper() and w4[0].isupper()
                    and w2 in {'of', 'the'}
                    and all(w.lower() not in stopwords for w in [w1, w3, w4])):
                    entidade = f"{w1} {w2} {w3} {w4}"
                    entidades.append(entidade)
                    i += 4
                    continue
            
            if i + 2 < n:
                w1, w2, w3 = palavras[i], palavras[i+1].lower(), palavras[i+2]
                if (w1[0].isupper() and w3[0].isupper()
                    and w2 in {'of', 'the'}
                    and w1.lower() not in stopwords and w3.lower() not in stopwords):
                    entidade = f"{w1} {w2} {w3}"
                    entidades.append(entidade)
                    i += 3
                    continue

            i += 1

    return entidades

texto = "The Sealers of New Londo were powerful. Age of Fire is important. Witch of Izalith is feared."
print(extrair_entidades_maiores(texto))
```

```python
def extrair_novas_entidades(texto):
    frases = re.split(r'[.!?;,]', texto)
    novas_entidades = []
    vistos = set()

    for frase in frases:
        palavras = frase.strip().split()
        n = len(palavras)
        i = 0

        while i < n:
            extraiu = False
            for tamanho in range(5, 1, -1):
                if i + tamanho - 1 < n:
                    grupo = palavras[i:i+tamanho]

                    if all(w[0].isupper() for w in grupo):
                        entidade_str = ' '.join(grupo)
                        entidade_lower = entidade_str.lower()

                        if entidade_lower not in conhecidos_lower and entidade_lower not in vistos:
                            novas_entidades.append(entidade_str)
                            vistos.add(entidade_lower)
                            i += tamanho
                            extraiu = True
                            break
            if not extraiu:
                i += 1

    return novas_entidades

```

```python
def unificar_dataframes(df1, df2):
    import numpy as np

    colunas_necessarias = ['id', 'alias']

    # Adiciona colunas ausentes como NaN no df1
    for col in colunas_necessarias:
        if col not in df1.columns:
            df1[col] = np.nan

    # Adiciona colunas ausentes como NaN no df2
    for col in colunas_necessarias:
        if col not in df2.columns:
            df2[col] = np.nan

    # Reordena colunas para facilitar concatenação (mantém ordem de df1)
    df2 = df2[df1.columns.intersection(df2.columns).tolist() + list(set(df1.columns) - set(df2.columns))]

    # Concatena os dois dataframes
    df_unificado = pd.concat([df1, df2], ignore_index=True)

    return df_unificado
```
